```python
#request 문을 이용한 url 파일 데이터 다운로드 
import requests 
url = 'http://www.google.com/' #google url 주소
google = requests.get(url).text  #requests.get(url) = url 주소 를 보내서 요청 
#get().text  = url 주소의 html text 를 가져온다

print(google) #html text


```

```python
#get(url).text 를 하면  문서 내용이 상당히 길기 때문에 어느 태그에 어떤 속성을 보고 싶을지 결정

import re 
re.findall('<div class=(.*?)</div>',google)
```

```python
#json 파일 데이터 보기 
import json #json file load library
from pandas.io.json import json_normalize #json 파일을 pandas 의 데이터 프레임으로 만드는 라이브러리 이다.

url = 'https://rank.search.naver.com/rank.js'
res = req.get(url)
json_normalize(json.loads(res.text),['data','data'])
#normalize = 데이터를 pandas 의 dataFrame 으로 만든다.
#json_normalize(json.loads(res.text),['data','data']) data data 는 불러올 데이터의 위치 를 말한다.


```

```python
#urllib request 
'''
urllib request vs requests
requests 라이브러리는 데이터를 받아오면 데이터는 dictionary 형태 이다.
urllib request 은 이진 바이너리 형태로 받아온다.
'''

import urllib.request
url = 'https://upload.wikimedia.org/wikipedia/commons/4/47/PNG_transparency_demonstration_1.png'
urllib.request.urlretrieve(url,'test.png') 
#urllib.request.urlretrieve(url,'test.png') 데이터를 바로 저장할수 있다.
```

```python
#urlopen 을 이용해 데이터를 파이썬에서 읽을수 있다.
import matplotlib.pyplot as plt
import matplotlib.image as mpimg #이미지 파일을 읽을수 있고 

png = urllib.request.urlopen(url).read() #데이터를 파이썬에서 바로 읽는다.

with open('test2.png','wb') as f: #png 파일은 urllib.request 라이브러리기 때문에 이진 바이너리 로 써야한다.
  f.write(png)

plt.imshow(mpimg.imread('test2.png')) 
#test2.png 파일을 읽어와 image show 함수 통해 파일에 저장된 png 이미지를 가져온다 
#mpimg.imread('파일의 이름') 해당 이름의 이미지 파일을 읽어오는 함수 
#plt.imshow(이미지 파일) 읽어온 이미지 파일을 보여준다.
```

```python
#query 파라미터 보내기
#기상청 데이터 가져오기
url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
parmas = urllib.parse.urlencode({'sidid':'109'})
print(parmas) #sidid = 109
'''
urllib.parse.urlencode(query)
query : {key:value}
url 주소에 들어갈 쿼리를 만들어서 저장
'''
url = url + "?" + parmas #데이터 쿼리문을 저장해서 url 을 다시 저장
print(url)
data = urllib.request.urlopen(url).read() #해당 url 을 읽어오고
text = data.decode("utf-8") #언어를 프로그래밍에서 읽어올 utf-8 로 읽어온다
text

```

```python
#tag 를 이용한 접근 
from bs4 import BeautifulSoup
html = """
<html>
  <body>
      <h1> a </h1>
      <p> b abc</p>
      <p> aedc </p>
    <body>
  </html>
"""

soup = BeautifulSoup(html,'html.parser') #soup 에 html tag를 저장 html.parser 를 쓰면 원시코드를 순수 문자열 객체를 해석
h1 = soup.html.body.h1 #태그를 이용해서 접근 한다.
p1 = soup.html.body.p
p2 = p1.next_sibling 
p3 = p1.next_sibling.next_sibling
print(p1)
print(p2)  #next_sibling 다음 데이터를 읽어 온다 p 태그 를 가져올려면 next_sibling  을 두번 쓴다.
print(p3)
```

```python
#id 를 사용해서 가져오기 
html = """
<html><body>
 <h1 id="title"> 파이썬으로 웹문서 읽기 </h1>
 <p id="body"> 페이지 분석기능 </p>
 <p> 페이지 정렬 </p>
</body></html>
"""
soup = BeautifulSoup(html,"html.parser")
title =soup.find(id = 'title')
body = soup.find(id = 'body')

print(title)
print(body)
```

```python
html = """
<html><body>
 <ul>
  <li><a href="http://www.naver.com">naver</a></li>
  <li><a href="http://www.daum.com">daum</a></li>
 </ul>
</body></html>
"""

soup = BeautifulSoup(html,'html.parser')
links = soup.find_all("a")
print(links) #리스트 형태로 a 태그 링크가 전부 들어간다.

for a in links: #각각 의 데이터로 접근 하고 싶을때 
  href = a.attrs["href"] #link a 에 href = href 주소 가 들어간다.
  text = a.string #a tag 의 text 가져오기 
  print(text,href)
```

```python
#dom 요소 파악 
soup.prettify() #각각의 태그 나 내용 의 요소 파악 


soup.a.attrs['href'] #href 의 속성 값에 접근
if 'href' in soup.a.attrs: #속성 여부 확인
  print(True)
```

```python
#기상청 데이터를 url open
url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
res = urllib.request.urlopen(url)

soup = BeautifulSoup(res, 'html.parser')
soup.prettify()
title = soup.find("title").string
wf = soup.find('wf').string #find 는 한번의 태그 만 가져옴

print(title)
print(wf)
```

```python
#copy select 를 이용해서 윤동주 시인의 시 가져오기 

import urllib.request
from bs4 import BeautifulSoup

#mw-content-text > div.mw-parser-output > ul:nth-child(6) > li > b > a
url = "https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC"
res = urllib.request.urlopen(url)
soup = BeautifulSoup(res,"html.parser")
atag_list = soup.select('#mw-content-text > div.mw-parser-output > ul > li > b  a')

for i in atag_list:
  print(i.attrs['title'])
```
