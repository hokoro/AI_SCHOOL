##유방암 셋을 통한 훈련 세트 와 테스트 세트의 성능 변화 


```python
#import 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics 
from sklearn.datasets import load_breast_cancer

#dataload
cancer = load_breast_cancer() #데이터 셋 가져오기 

print(cancer.keys())
```

```python
print(cancer.data)
print('--------------------------------------------------')
print(cancer.target)
print('--------------------------------------------------')
print(cancer.data.shape)
print('--------------------------------------------------')
print(cancer.target.shape)
print('--------------------------------------------------')
print(cancer.target_names) #악성 양성 
print('--------------------------------------------------')
print(cancer.feature_names)
```

```python
#양성 과 악성을 수를 세서 정리 
import numpy as np 

for n,v in zip(cancer.target_names ,np.bincount(cancer.target)):
  print({n:v}) #양성과 악성의 수를 디셔너리 형태로 출력
  #np.bincount(data) 데이터의 수를 종류 별로 카운팅 
  
```

```python
#train_test_split 
x = cancer.data
y = cancer.target

x_train,x_test,y_train,y_test = train_test_split(x,y,stratify = y,random_state = 0)
#stratify = y y의 종류가 한쪽으로 편향 안돼고 골고루 섞어지게 만드는 매개변수

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
train_acc = []
test_acc = []

n_neighbors = range(1,11)
for n in n_neighbors:
  model = KNeighborsClassifier(n_neighbors=n)
  model.fit(x_train,y_train)
  predict = model.predict(x_test)
  score = metrics.accuracy_score(y_test,predict)

  print(score)
  '''
  train_acc.append(model.score(x_train,y_train)) #질문
  test_acc.append(model.score(x_test,y_test))
  '''
```

```python
#model 시각화 

plt.plot(n_neighbors,train_acc,label = 'train_score')
plt.plot(n_neighbors,test_acc,label = 'test_score')
plt.xlabel('neighbors')

plt.ylabel('accuracy')
plt.legend()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split 
import numpy as np
X,y = mglearn.datasets.make_wave(n_samples = 9) #곡선 모양으로 만들어진 데이터 
x_train,x_test,y_train,y_test = train_test_split(X,y,random_state = 0)

model = KNeighborsRegressor(n_neighbors = 3,weights='distance')
model.fit(x_train,y_train)

pre = model.predict(x_test)
print(pre)
print(y_test)

Mse = np.sum(np.square(y_test - pre))/len(y_test)

print(Mse) # 0 에 가까워 져야 좋음 오차가 존재 하지 않으니까.
```

##영화 데이터 회귀 로 풀기 
###distance 가중치 방식을 사용 

```python
from sklearn.neighbors import KNeighborsRegressor

# 영화에 대한 3개의 특성
X_train = [
  [0.5, 0.2, 0.1],
  [0.9, 0.7, 0.3],
  [0.4, 0.5, 0.7]
]
# 영화에 대한 별점
y_train = [5.0, 6.8, 9.0]

model = KNeighborsRegressor(n_neighbors=3,weights= 'distance')
'''
uniform 방식을 사용 했을때 거리 평균으로 전부 통일 이된다.
[6.93333333 6.93333333 6.93333333]
distance 방식을 사용 하면 각각의 값에 따라 가중치 가 부여 되기 때문에 값이 다 달라진다.
[7.28143288 7.76451922 6.8457845 ]
'''
model.fit(X_train,y_train)


# 새로운 영화 3건에 대해 별점을 예측
X_test = [
  [0.2, 0.1, 0.7],
  [0.4, 0.7, 0.6],
  [0.5, 0.8, 0.1]
]
predict = model.predict(X_test)
print(predict)
```
