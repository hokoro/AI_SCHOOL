```python
#import pandas ,numpy 

import pandas as pd 
import numpy as np 
df = pd.read_csv('/content/faults.csv')
```

```python
#data 확인 
df.info()
```

```python
#데이터를 좀더 간편화 시킨다.
df_dataset = df[['X_Minimum','X_Maximum','Steel_Plate_Thickness','LogOfAreas',
                 'Pastry','Z_Scratch','Bumps']]
print("df_dataset.shape : ", df_dataset.shape)
print(df_dataset.info())
```

```python
#train /test split
df_dataset = np.asarray(df_dataset, dtype = 'float32')

df_dataset_x = df_dataset[0:1,:-3] #독립변수
df_dataset_y = df_dataset[0:1,-3:] #종속변수

print(df_dataset_x.shape)
print(df_dataset_x)
print(df_dataset_y.shape)
print(df_dataset_y)
```

```python
#정규 분포 평균 과 표준 편차 ,입력과 출력 크기 model parameter 정의
RND_MEAN = 0 
RND_STD = 1

input_cnt = df_dataset_x.shape[-1]
output_cnt = df_dataset_y.shape[-1]

weight = np.random.normal(RND_MEAN,RND_STD ,size = [input_cnt,output_cnt])
bias = np.random.normal(RND_MEAN,RND_STD ,size = [output_cnt])

print(weight)
print("------------------------")
print(bias)
```

```python
p_1 = np.matmul(df_dataset_x[0],weight[:,0]) + bias[0]
p_2 = np.matmul(df_dataset_x[0],weight[:,1]) + bias[1]
p_3 = np.matmul(df_dataset_x[0],weight[:,2]) + bias[2]

print(p_1,p_2,p_3)
result = np.matmul(df_dataset_x,weight) + bias
print(result)
```

```python
RND_MEAN = 0 
RND_STD = 1

input_cnt = df_dataset_x.shape[-1]
output_cnt = df_dataset_y.shape[-1]
print(input_cnt)
print(output_cnt)
weight = np.random.normal(RND_MEAN,RND_STD ,size = [input_cnt,output_cnt])
bias = np.random.normal(RND_MEAN,RND_STD ,size = [output_cnt])

print(weight)
print("------------------------")
print(bias)
```

```python
#학습 데이터 셋과 테스트 셋 분류 
mb_size= 2 # 미니 배치 사이즈 
train_ratio = 0.8 #분리 비율 train set 이 80퍼센트 정도 나타 낸다.

print(df.shape)
print(df.shape[0] * train_ratio)

mb_step_count = df.shape[0] * train_ratio // mb_size #MB1 MB2 MB3 MB4
print(mb_step_count)

```

```python
#test set index 
test_begin_index = mb_size * mb_step_count
print(test_begin_index)

#전체 데이터 셔플 
shuffle_map = np.arange(df.shape[0])
print(shuffle_map)
np.random.shuffle(shuffle_map)

print(shuffle_map)
```

```python
#미니 배치 사이즈를 shuffle map numpy 를 shuffle 하여 index 별로 나눔
mb_data_1 = df[shuffle_map[0:2]]
print(mb_data_1)
mb_data_2 = df[shuffle_map[2:4]]
print(mb_data_2)
mb_data_3 = df[shuffle_map[4:6]]
print(mb_data_3)
mb_data_4 = df[shuffle_map[6:8]]
print(mb_data_4)
```

```python
#독립변수 종속 변수 분할 
mb_data_1_x = mb_data_1[:,:-3]
mb_data_1_y = mb_data_1[:,-3:]
print(mb_data_1_x)
print(mb_data_1_y)
mb_data_2_x = mb_data_2[:,:-3]
mb_data_2_y = mb_data_2[:,-3:]

print(mb_data_2_x)
print(mb_data_2_y)
mb_data_3_x = mb_data_3[:,:-3]
mb_data_3_y = mb_data_3[:,-3:]

print(mb_data_3_x)
print(mb_data_3_y)
mb_data_4_x = mb_data_4[:,:-3]
mb_data_4_y = mb_data_4[:,-3:]

print(mb_data_4_x)
print(mb_data_4_y)
```

```python
#mb_yhat predict 값 구하기 
mb1_yhat0 = np.matmul(mb_data_1_x,weight[:,0]) + bias[0]
mb1_yhat1 = np.matmul(mb_data_1_x,weight[:,1]) + bias[1]
mb1_yhat2 = np.matmul(mb_data_1_x,weight[:,2]) + bias[2]
print("MB1")
print(mb1_yhat0)
print(mb1_yhat1)
print(mb1_yhat2)


mb2_yhat0 = np.matmul(mb_data_2_x,weight[:,0]) + bias[0]
mb2_yhat1 = np.matmul(mb_data_2_x,weight[:,1]) + bias[1]
mb2_yhat2 = np.matmul(mb_data_2_x,weight[:,2]) + bias[2]

print("MB2")
print(mb2_yhat0)
print(mb2_yhat1)
print(mb2_yhat2)

mb3_yhat0 = np.matmul(mb_data_3_x ,weight[:,0]) + bias[0]
mb3_yhat1 = np.matmul(mb_data_3_x,weight[:,1]) + bias[1]
mb3_yhat2 = np.matmul(mb_data_3_x,weight[:,2])+ bias[2]

print("MB3")
print(mb3_yhat0)
print(mb3_yhat1)
print(mb3_yhat2)

mb4_yhat0 = np.matmul(mb_data_4_x ,weight[:,0]) + bias[0]
mb4_yhat1 = np.matmul(mb_data_4_x,weight[:,1]) + bias[1]
mb4_yhat2 = np.matmul(mb_data_4_x,weight[:,2])+ bias[2]

print("MB4")
print(mb4_yhat0)
print(mb4_yhat1)
print(mb4_yhat2)

```
