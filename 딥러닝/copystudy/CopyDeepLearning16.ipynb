```python
import numpy as np

x_list = [1000,0,-1000]
def sigmoidplus(x):
  y = 1/(1+np.exp(-x))
  return np.round(y,3)

for x in x_list:
  print("x:{} exp:{}".format(x,sigmoidplus(x)))

'''
x:1000 exp:1.0
x:0 exp:0.5
x:-1000 exp:0.0
/usr/local/lib/python3.7/dist-packages/ipykernel_launcher.py:5: RuntimeWarning: overflow encountered in exp
음수는 오버 플로우 문제가 발생한다 
'''

```

```python
import numpy as np 

x_list = [1000,0,-1000]
def sigmoidminus(x):
  y = np.exp(x)/(1+np.exp(x))
  return np.round(y,3)


for x in x_list:
  print("x:{} exp:{}".format(x,sigmoidminus(x)))

'''
x:1000 exp:nan
x:0 exp:0.5
x:-1000 exp:0.0
/usr/local/lib/python3.7/dist-packages/ipykernel_launcher.py:5: RuntimeWarning: overflow encountered in exp
  """
/usr/local/lib/python3.7/dist-packages/ipykernel_launcher.py:5: RuntimeWarning: invalid value encountered in double_scalars
  """
양수 값에서 값이 무한대에 가까워지는 overflow 문제 가 발생
'''
```

```python
##일반화 
import numpy as np
x = [1000,0,-1000]

def sigmoidtotal(x):
  y = np.exp(-np.maximum(-x,0)) /(1+np.exp(-np.abs(x)))
  return y

for x in x_list:
  print("x:{},exp:{}".format(x,sigmoidtotal(x)))

'''
일반화 를 시키니까 overflow 문제를 해결 
'''
```

```python
##plus 버전 
x_list = [1000,0,-1000]
z_list = [1,0]
import numpy as np

def sig_cross_entropyplus(x,z):
  H= x - x*z + np.log(1+np.exp(-x))
  return np.round(H,3)


for z in z_list:
  for x in x_list:
    print("z:{} , x:{} ,H:{}".format(z,x,sig_cross_entropyplus(x,z)))

```

```python
##x minus 버전 
x_list = [1000,0,-1000]
z_list = [1,0]

def sig_cross_entropyminus(x,z):
  H = -x*z + np.log(1+np.exp(x))
  return H

for z in z_list:
  for x in x_list:
    print("z:{} , x:{} ,H:{}".format(z,x,sig_cross_entropyminus(x,z)))
```
