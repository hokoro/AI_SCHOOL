```python
#class 를 통한 초기화 + y 예측 값을 임의로 넣어 완성 해보기 
class Y_pred():
  def __init__(self,theta_0,theta_1):
    self.theta_0 = theta_0
    self.theta_1 = theta_1
  

def out_y_hat(self,input_x):
  y_hat_row = []
  for i in range(len(input_x)):
    y_hat = self.theta_0 + (self.theta_1 * input_x[i]) #input 값별로 y 예측 데이터 저장 
    y_hat_row.append(y_hat) #값마다 list 로 저장
  return y_hat_row #return  


Y_pred.y_hat = out_y_hat #클래스 외부에서 함수 선언 하기 위한 방법
```

```python
x = [580,700,810,840]
Y_pred_C = Y_pred(theta_0=1,theta_1=2) #y_pred 클래스 객체 선언 후 초기화 
Y_pred_C.y_hat(input_x = x)  #x 데이터를 예측 함수 실행
```

```python
#theta0 theta1 학습 하기 
#데이터 및 변수 초기화
import numpy as np 
  
x = np.asarray([580,700,810,840])
y = np.asarray([374,385,375,401])

RND_Mean = 0
RND_Std = 1

input_cnt = 1
output_cnt = 1




```

```python
#메인 실행문
def main_execute(x,y,epochs,report,lr):
  model_init()
  sse_row, theta0_row, theta1_row =run_train(x,y,epochs,lr,report)

  return sse_row,theta0_row,theta1_row
```

```python
#모델 초기화 = 가중치(theta1) + bias(theta0)
def model_init():
  global theta0 ,theta1
  theta0 = np.random.normal(RND_Mean,RND_Std,size = [output_cnt])
  theta1 = np.random.normal(RND_Mean,RND_Std,size = [input_cnt,output_cnt])
```

```python
#함수의 예측값 만들기 
def forward_neuralnet(x):
  y_hat = theta0 + theta1 * x
  return y_hat
```

```python
#목적 함수 계산 
def forward_postproc(y_hat,y):
  diff = y_hat - y
  square = np.square(diff)
  sse_sum = np.sum(square)
  sse = sse_sum * 1/2
  return sse
```

```python
def back_propagation(lr,y_hat):
  global theta0 ,theta1
  theta0 = theta0 - lr * (np.sum(y_hat - y))
  theta1 = theta1 - lr * (np.sum(y_hat - y) * x)
```

```python


def run_train(x, y, epoch_count, lr, report):
    print("Initial theta_0 : {}".format(theta0))
    print("Initial theta_1 : {}".format(theta1))
    sse_row = []
    theta_0_row, theta_1_row = [], []

    print(epoch_count)
    for epoch in range(epoch_count):
        y_hat = forward_neuralnet(x)
        sse = forward_postproc(y_hat, y)

        sse_row.append(sse)

        theta_0_row.append(theta0)
        theta_1_row.append(theta1)
        
        back_propagation(lr, y_hat)

        if report > 0 and epoch % report == 0:
            print("Epoch - {}".format(epoch + 1))
            print("SSE : {}".format(sse))
    print("=============================")
    print("Final SSE : {}".format(sse))

    return sse_row, theta_0_row, theta_1_row

sse_row, theta_0_row, theta_1_row = main_execute(x,y,100,2,0.0000001)
  
```

```python
import matplotlib.pyplot as plt

plt.plot(sse_row,'--o',color = 'blue')
plt.xlabel('epoch')
plt.ylabel('sse')
plt.grid()
plt.show()
```
