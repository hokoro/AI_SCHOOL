```python
from pykospacing import Spacing
from hanspell import spell_checker
spacing = Spacing()
```

```python
comment_list = []
for comment in youtube_comments[3:]:
  comment_list.append(spacing(comment.text))
```

```python
comment_list
```

```python
from ckonlpy.tag import Twitter


twi = Twitter()
#twi.add_dictionary("Î≥ÑÏÇ¨Ïû•",'Noun')
twi.add_dictionary("Î≥ÑÏùÄ",'Noun')
twi.add_dictionary("ÌíÄÎ≤ÑÏ†Ñ","Noun")
twi.add_dictionary("Ïò•Íµ¨Ïä¨",'Noun')
twi.add_dictionary("Ïò§ÎßàÏù¥Í±∏",'Noun')
```

```python
comment_nouns = []

for comment in comment_list:
  comment_nouns.append(twi.nouns(comment))
  
```

```python
no_meaning = ['Ïñ¥Ïöî','ÏßÑÏßú','Ï†ïÎßê','Ïò§Îäò','Ìï≠ÏÉÅ','ÏÉùÍ∞Å']
tokens = []
for comment in comment_nouns:
  for token in comment:
    if len(token) >1:
      tokens.append(token)
```

```python
tokens2 = []

for tk in tokens:
  if tk not in no_meaning:
    tokens2.append(tk)
```

```python
import pandas as pd

df = pd.Series(tokens2)
```

```python
token_count = df.value_counts().sort_values(ascending = False)
token_count
```

```python
word_dict = {}
for index,count in zip(token_count.index , token_count.values):
  word_dict[index] = count
```
