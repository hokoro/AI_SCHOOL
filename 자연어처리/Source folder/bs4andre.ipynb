```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
url = 'http://www.pythonscraping.com/pages/page3.html'

req = requests.get(url)
html= req.text
print(html)

soup = BeautifulSoup(html,'html.parser')
```

```python
img_list = soup.findAll('img',{'src':re.compile('\.\.\/img\/gifts/img.*\.jpg')})

img_list_total =[]
for img in img_list:
  #print(img)
  img_list_total.append(re.findall('[a-z0-9]+.[a-z]+',str(img)))
```

```python
print(img_list_total)
```

```python
answer_total = []
for img in img_list_total:
  answer_total.append(img[2])
```

```python
for i in answer_total:
  print(i,end = '\n')  
```
