````python
!pip3 install requests
!pip3 install bs4
import requests
from typing import List
from bs4 import BeautifulSoup  # html 파싱을 위한 라이브러리
import re # 정규표현식 사용을 위한 라이브러리


# 다음의 웹페이지에서, 상품에 해당하는 이미지의 고유 아이디를 파싱해주세요!
URL = "http://www.pythonscraping.com/pages/page3.html"
````

````python
pattern = re.compile(r'^123')
````

````python
# requests 라이브러리를 활용하면 한줄로 html을 간단하게 로드할 수 있습니다!
html = requests.get(URL).text
soup = BeautifulSoup(html, "html.parser")

#### TODO ####
# 나머지는 여러분의 몫입니다 :)
# hint: soup.findAll(), re.compile()를 활용해보세요!.
# 참고로 tag['src']로 attriubute를 접근할 수 있습니다.
img_list = soup.findAll('img',{'src':re.compile('img(\d+).jpg$')})

print(img_list)
# 다시 말하지만, "상품"에 해당하는 이미지의 id만 가져오면 됩니다~
img_ids: List[str] = [re.search('img\d+\.jpg', img['src']).group(0) for img in img_list]
##############

````

````python
img_list = soup.findAll('img',{'src':re.compile('\.\.\/img\/gifts/img.*\.jpg')}) #로고 만 빼고 가져오기 
img_list_total =[]
for img in img_list:
  #print(img)
  img_list_total.append(re.findall('img\d+\.jpg',str(img))) 
answer_total = []
for img in img_list_total:
  answer_total.append(img)
for i in answer_total:
  print(i,end = '\n')  
````
