##정규표현식 

- "a.b" :a와 b 사이에 매치되는 문자가 1개 존재한다.

- "a*b" 을 이용한 문자 매치 :0회부터 무한대 까지 반복 가능하다 

- "a+b" 을 이용해 문자 매치가 1회 이상 있을 경우 

- "a?b" 은 문자가 0회 또는 1회 만 매치 

- "a{m}b" 은  바로 앞에 있는 문자가 m 회 매칭 된다는 의미 

```python
import re
```

```python
tokens = ["ct","cat","caat"]
```

```python
#. 을 이용한 메타 문자 표현 .중심으로 앞뒤 글자 가 있고 그사이에 메타 문자가 존재할경우
my_expression = re.compile("c.t")

for token in tokens:
  if my_expression.search(token):
    print(token," yes")
  else:
    print(token," No")

```

```python
#* 을 이용한 문자 매치 :0회부터 무한대 까지 반복 가능하다 
tokens = ["ct","cat","caat","caaaaaat"]

my_expression = re.compile("ca*t")

for token in tokens:
  if my_expression.search(token):
    print(token+" yes")
  
  else:
    print(token+" No")
```

```python
#+ 을 이용한 문자 매치 1회 이상 반복 
tokens = ["ct","cat","caat","caaaaaat"]

my_expression = re.compile("ca+t")

for token in tokens:
  if my_expression.search(token):
    print(token+" yes")
  
  else:
    print(token+" No")
```

```python
#? 을 이용한 문자 매치 0회 또는 1회 출현 
tokens = ["ct","cat","caat","caaaaaat"]

my_expression = re.compile("ca?t")

for token in tokens:
  if my_expression.search(token):
    print(token+" yes")
  
  else:
    print(token+" No")
```

```python
#{m} 을 이용한  바로 앞에 있는 문자 매치 가 m 회 반복 된다는 의미  
tokens = ["ct","cat","caat","caaat","caaaaaat"]

my_expression = re.compile("ca{2}t")

for token in tokens:
  if my_expression.search(token):
    print(token+" yes")
  
  else:
    print(token+" No")
```

```python
#{m,n} 바로 앞에 문자가 m 회 이상 n 회 이하 반복 
tokens = ["ct","cat","caat","caaat","caaaaaat"]

my_expression = re.compile("ca{2,5}t")

for token in tokens:
  if my_expression.search(token):
    print(token+" yes")
  
  else:
    print(token+" No")
```

```python
# ^ 문자열의 맨 앞부분이 일치함을 의미 
texts = ["AI is difficult","this is AI"]


my_expression  = re.compile("^AI")

for text in texts:
  if my_expression.search(text):
    print(text+" ->yes")

  else:
    print(text+" ->No")
```

```python
#$ 문자열이 끝부분의 단어와 일치 
texts = ["Python is easy","Is this Python"]

my_expression = re.compile("Python$")

for text in texts:
  if my_expression.search(text):
    print(text+" -> yes")

  else:
    print(text+" -> No")
```

```python
#A|B A 또는 B 가 text 안에 있다. 
texts = ["I hate you","I love you","I like you"]

my_expression = re.compile("love|hate")

for text in texts:
  if my_expression.search(text):
    print("yes")
  else: 
    print("No")
```
