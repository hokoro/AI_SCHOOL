###GeoJson 실습

```python
import json #json 파일 관리 라이브러리
import pandas as pd  #pandas 
import folium #지도 관리 라이브러리
from folium import plugins 
from ipywidgets import interact # Ineractive 시각화 기능.
from sklearn.preprocessing import MinMaxScaler #min max 최적화 라이브러리 
```

## 1.GeoJson 파일을 읽어와서 탐색한다.

```python
f = open('/content/drive/MyDrive/인공지능사관학교/빅데이터/04_시각화와 분석_업데이트 2/data/geo_seoul.json',encoding='utf8')
get_seoul = json.load(f) #json 파일 load 함수  json.load(filename)
f.close()
```

##get_seoul 정보 
- id : 지역구 이름 = name : 지역구 이름 
- name_eng : 지역구 영어 이름 
- { geometry :{coordinates : [[[]]]}} 지리 정보를 담고 있다


```python
#geo seoul 
get_seoul['features'][0]
```

```python
get_seoul['features'] #전체 특성 가져오기 
```

```python
get_seoul['features'][0]['id'] #feature list(index) 안에  지리정보 dict 가 들어가있다 
```

```python
#구정보를 가져 와야 한다 
kus = []
for a_dict in get_seoul['features']:
  kus.append(a_dict['id'])


print(kus)
```

```python
#지리 정보 가 필요 
get_seoul['features'][0]['geometry']
```

```python
#지리 타입 정보 
types = []
for a_dict in get_seoul['features']:
  types.append(a_dict['geometry']['type'])


print(types)
```

```python
#경계 좌표 가져오기 
coords =[]

for a_dict in get_seoul['features']:
  coords.append(a_dict['geometry']['coordinates'][0])

coords
```

```python
my_coords2 =  [[ [a_lat, a_long] for (a_long, a_lat) in a_polygon ] for a_polygon in coords ] # 위도 경도 바꾸기 folium 에서는 위도 경도 로 입력해야함 

my_coords2[0]
```

```python
# 구의 경계 좌표를 사용해서 Polygon으로 출력해 본다.
my_map=folium.Map(location = [37.55, 127],  tiles="Stamen Toner", zoom_start=10, width='50%', height='80%')
for a_ku, a_coords in zip(kus, my_coords2):
    folium.Polygon(locations=a_coords,#위치정보        
                   fill=True,#색깔을 전부 칠할것인디
                   fill_color='turquoise', #바탕 색
                   fill_opacity=0.3,#불투명도
                   color="red", #라인 색깔
                   weight=1,#라인 두께
                   opacity=0.5, #경계선 불투명도 
                   tooltip=a_ku #
                   ).add_to(my_map) 
my_map
```

##3 Chorpleth Map

3.1 데이터 전처리:

```python
# 인구 정보 DataFrame을 읽어 온다.
my_data = pd.read_csv("/content/drive/MyDrive/인공지능사관학교/빅데이터/04_시각화와 분석_업데이트 2/data/data_seoul_population.csv",header="infer",encoding="utf8")
my_data.head(3)
```

```python
#구는 index 로 들어가야 함 + min max scaler 를 해야함 why ? -> 이따가 지도 에 표시할떄 퍼센트 단위롤 들어가는 소수점이 제일 베스트 이다. 
my_columns = ['세대수','인구','남자','여자']
my_scaler = MinMaxScaler()
X_scaled = my_scaler.fit_transform(my_data[my_columns])

my_data_scaled = pd.DataFrame(data = X_scaled,columns = my_columns,index = my_data['구'])
my_data_scaled
```

```python
my_data_scaled['세대수'].index
```

```python
my_data_scaled.index
```

```python
get_seoul
```

3.2 Choropleth map 출력 

```python
my_map = folium.Map(location = [37.55,127],tiles = 'Stamen Toner',zoom_start = 10,width = '50%',height = '80%')
folium.Choropleth(geo_data = get_seoul, #지리정보 가 담은 데이터
                  data = my_data_scaled['인구'], #색깔을 적용시킬 data (0~1 사이에)
                  fill_opacity = 0.5, #불투명도
                  fill_color = 'BuPu',#영역 안에 색깔 
                  line_weight = 1,#라인의 크기
                  line_color = 'blue',#영역 라인의 색깔 
                  line_opacity = 0.4, #라인의 투명도 
                  legenend_name = 'Population',#데이터 의 수치를 막대화 
                  key_on = 'feature.id' #특성의 id 를 key 로 
                  ).add_to(my_map)
my_map
```

```python
#세대수 로 채색 
my_map = folium.Map(location = [37.55,127],tiles = 'Stamen Toner',zoom_start = 10,width = '50%',height = '80%')
folium.Choropleth(geo_data = get_seoul,
                  data = my_data_scaled['세대수'],
                  fill_opacity = 0.5,
                  fill_color = 'YlGn',
                  line_weight = 0.5,
                  line_color = 'orange',
                  line_opacity = 0.5,
                  legend_name = 'Households',
                  key_on = 'feature.id').add_to(my_map)
my_map
```

##4.interactive 시각화 

```python
folium.Choropleth()
```

```python
my_columns = ['세대수','인구','남자','여자']
my_color = ['BuGn', 'BuPu','GnBu', 'OrRd', 'PuBu', 'PuBuGn', 'PuRd', 'RdPu','YlGn', 'YlGnBu','YlOrBr','YlOrRd']
my_tiles = ['OpenStreetMap' ,'Stamen Terrain' ,'Stamen Toner','Stamen Watercolor' ,'CartoDB Positron' ,'CartoDB Dark_matter']

@interact(data_column = my_columns,color = my_color,tiles = my_tiles,zoom = (8,12))
def choropleth_map(data_column = '세대수',color='BuGn',zoom = 10):
   my_map = folium.Map(location = [37.55,127],tiles = 'Stamen Toner',zoom_start =zoom,width = '50%',height = '80%')
   folium.Choropleth(geo_data = get_seoul,data = my_data_scaled[data_column],fill_opacity = 0.5,fill_color = color,line_weight = 0.5,line_color = 'orange',line_opacity = 0.5,key_on = 'feature.id').add_to(my_map)
   return my_map
 
```

```python
res = interact(choropleth_map,data_column = my_columns , color = my_color,tile = my_tiles,zoom = (8,12))
```
