```python
import konlpy
```

####꼬꼬마 분석기 

```python

kkma = konlpy.tag.Kkma()
my_text = "한국어를 사용해 봅시다 너무 재미있어요 함께 분석해 볼까요?"
```

```python
#문장 분절 띄워쓰기
kkma.sentences(my_text)
```

```python
#형태소 단위 분절 
kkma.morphs(my_text) #분석디마다 이상함
```

```python
#명사 추출 
kkma.nouns(my_text)
```

```python
#품사태깅  = 단어가 어떤 품사 인지 
kkma.pos(my_text)
```
