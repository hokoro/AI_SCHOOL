```python
import numpy as np 
import scipy.stats as st
```

```python
x = np.array([25,24,24,27,29,31,28,24,25,26,25,18,30,28,23,26,27,23,16,20,22,22,25,24,24,25,25,27,26,30,25,25,26,26,25,24])
n = len(x)
print(n)
```

SEM = $ s \over{\sqrt{n}} $을 계산해 둔다.

s = 표본 표준 편차 
n = 표본의 크기 

```python
x_mean = x.mean()
SEM = x.std(ddof=1)/np.sqrt(n)
```

90% 신뢰 구간 

```python
alpha = 0.1
```

```python
# 정확한 Student t 분포의 분위수 사용.
(x_mean + st.t.ppf(alpha/2,df=n-1)*SEM, x_mean + st.t.ppf(1-alpha/2,df=n-1)*SEM)
```

```python
#st.t.Interval = 구간을 구해주는 함수 
st.t.interval(alpha = 1- alpha ,df = n-1 , loc = x_mean , scale = SEM)
```

95% 신뢰 구간


```python
alpha = 0.05
```

```python
(x_mean + st.t.ppf(alpha/2,df=n-1)*SEM, x_mean + st.t.ppf(1-alpha/2,df=n-1)*SEM)
```

```python
st.t.interval(alpha = 1-alpha , df = n-1 , loc = x_mean,scale = SEM)
```

99% 신뢰구간

```python
alpha = 0.1
```

```python
x_mean + st.t.ppf(alpha/2 , df = n-1) * SEM , x_mean + st.t.ppf(1-alpha/1,df =n-1) *SEM
```
