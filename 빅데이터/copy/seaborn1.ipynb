##seaborn1

```python
import seaborn as sns 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load_dataset('데이터의 종류')
df = sns.load_dataset('mpg')
df.head(5)
```

##히스토그램  + KDE 

```python
#히스토 그램 
fig  = sns.histplot(data = df,x = 'mpg',bins = 50,color='turquoise')
fig.figure.savefig('output.png') #히스토 그램 외부로 저장 

'''
x = mpg column data 
bins = 계급 = x 축의 나눈 수 
color = 막대 색깔
'''
```

```python
#displot 
fig  = sns.displot(data = df,x = 'mpg',kde=False,rug=False,bins=50,color = 'red',kind='hist')
'''
kind 화면에 보여줄 시각화 종류를 고른다 
'''
```

```python
fig = sns.displot(data = df,x = 'mpg',kde = True,rug = False,bins = 50,color = 'red',kind = 'hist')
'''
kde = True 이면 막대 표에 선을 그리면서 나감 
근데왜 고점을 따라가는 형식이 아닌가? No 평균을 따라가는 형식 
'''
```

```python
#히스토그램  + rug 
fig = sns.displot(data=df,x = 'mpg',kde = False,rug = True,bins = 50,color = 'blue',kind = 'hist')
'''
rug = True = 관측치의 x축에 눈금 표시가 된다.
'''
```

```python
#kde 
fig  = sns.displot(data = df , x = 'mpg',hue = 'origin',kind = 'kde',aspect=1.5,height=3)

'''
hue = column(명목형) = 명목형 컬럼의 종류 별로 표에 색깔을 지정한다 
aspect =  표에 폭을 조절 
heigth =  표에 높이를 조절 
'''
```

##산점도 + 회귀선 

```python
fig  = sns.jointplot(data = df,x = 'weight',y = 'mpg',color = 'red',kind = 'scatter')
'''
x = data 의 weight 
y = data 의 mpg 
kind plt 에 보여주는 방식 
'''
```

```python
fig = sns.jointplot(data = df,x = 'weight',y = 'mpg',color = 'orange',kind = 'reg')
'''
reg = 회귀 선을 그어준다 
'''
```

```python
fig  = sns.jointplot(data = df , x = 'weight',y = 'mpg',color = 'green',kind = 'hex')
'''
hex 육각형 모델 로 표시 
'''
```

## 막대 그림 

```python
fig = sns.barplot(data = df,x = 'origin',y = 'mpg',ci = False)
```

```python
fig = sns.barplot(data = df, x = 'origin',y = 'mpg',estimator=np.std,ci = None)
```

```python
fig = sns.countplot(data = df,x = 'origin',palette = 'coolwarm')
```

```python
fig = sns.countplot(data = df,x = 'origin',hue = 'cylinders',palette = 'coolwarm')
'''
hue 해당 column 의 종류 별 데이터를 다 보여줌 
'''
```

## 상자 그림 바이올린 그림 




```python
#세로 방향 상자그림 
fig = sns.boxplot(data = df,y = 'mpg',color = 'green')
```

```python
fig  = sns.boxplot(data = df,x = 'mpg',color = 'green')
```

```python
#다중 상자 그림 notch = True
fig = sns.boxplot(data = df,x = 'origin',y = 'mpg',palette='coolwarm',notch = True)
```

```python
fig = sns.boxplot(data=df, x='mpg', y='origin', palette='coolwarm', notch=True)
```

```python
fig = sns.boxplot(data = df,x = 'origin',y = 'mpg',palette = 'coolwarm')
```

```python
#바이올린 그림 인자는 똑같음 
fig = sns.violinplot(data=df, x='origin', y='mpg', palette='coolwarm')
```

```python
#다중 바이올린 그림 
fig = sns.violinplot(data = df,x = 'cylinders',y = 'horsepower',pallette = 'coolwarm')
```

```python
#Swarm plot 
fig = sns.swarmplot(data = df,x = 'cylinders',y = 'horsepower',palette = 'coolwarm',size =2)
```

```python
# Strip plot.
fig = sns.stripplot(data=df, x='cylinders', y='horsepower', palette='coolwarm', size=3)
# fig.figure.savefig("output.png")
```

```python
# Violin plot + Smarm plot.
sns.violinplot(data=df, x='origin', y='horsepower')
fig = sns.swarmplot(data=df, x='origin', y='horsepower', color='black', size=3)
# fig.figure.savefig("output.png")
```
