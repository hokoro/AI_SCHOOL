```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

##산점도 행렬

```python
df = sns.load_dataset('iris')
df.head(5)
```

```python
df.shape
```

```python
fig = sns.pairplot(data = df,height = 1.5)
```

```python
fig = sns.pairplot(data = df,hue = 'species',height = 1.5)
```

```python
fig = sns.pairplot(data = df,hue='species',height=1.5,palette='coolwarm')
```

## 혼합 유형의 시각화 행렬 

```python
g = sns.PairGrid(data = df,height = 1.5) #PairGrid 객체 생성
fig = g.map(plt.scatter)  #g.map(표현 모양)
```

```python
g = sns.PairGrid(data = df,height = 1.5)
g.map_diag(sns.histplot,color = 'orange') #map 의 대각선 
g.map_upper(plt.scatter,color = 'green') #map 의 위쪽 부분
fig = g.map_lower(sns.kdeplot,color = 'blue') #대각선의 아래부분 kdeplot 아래부분 
```

## 다중 시각화 

```python
g = sns.FacetGrid(data = df,col = 'species',hue = 'species') #종속변수를 y축으로
fig = g.map(sns.histplot,'petal_length') #해당컬럼에 표현 방식을 종속변수 별로 보여준다 
```

```python
# 다중 히스토그램.
g = sns.FacetGrid(data=df, row='species', hue='species' ) #row를 기준으로 출력 
fig = g.map(sns.histplot, 'petal_length')
# fig.savefig("output.png")
```

```python
# 다중 산점도.
g = sns.FacetGrid(data=df, col='species', hue='species')
fig = g.map(plt.scatter, 'petal_length', 'petal_width') 

'''
sns.FacetGrid(data = df,col = 'species',hue = 'species') 종속변수 별 분류 기준 
g.map(표현 방식 ,x,y)
'''
# fig.savefig("output.png")
```
