### 1.불량 볼트 

```python
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np
n = 100 #시도 횟수
p = 0.02 #확률

# 0개의 볼트 불량 
print(np.round(st.binom.pmf(0,n,p),3)) 

#1개의 볼트 불량 
np.round(st.binom.pmf(1,n,p),3)

#2개의 볼트 불량 
np.round(st.binom.pmf(2,n,p),3)
```

### 2.손흥민


```python
p = 0.2

#5번 의 유효슈팅 중 1득점 이상 할 확률 
n= 5
1 - np.round(st.binom.pmf(0,n,p),3)
#10 게임에 매 게림 3회 유효슈팅이 있다고 가정하면 총 골수 
#10 * 3 = 30 
n = 30
mu = p * n
q = 1- p 
sigma = (n*p*q) ** 0.5

print(mu)
print(sigma)
```

### 3.학생 체중 

```python
n = 200

mu = 100
sigma  = 2 

#60 70 사이의 인원수 
n* np.round(st.norm.cdf(70,mu,sigma) - st.norm.cdf(60,mu,sigma),1)
#체중이 85kg 이상인 학생수 
print((1 - st.norm.cdf(,mu,sigma)) * n)

#90 % 보다 무거운 학생의 몸무게 
st.norm.ppf(0.9,mu,sigma) # 80kg
```

###4.카이제곱

```python
k = 9
#평균의 위치를 구하시오 
avg = k
print(avg) #평균과 자유도는 같다 

#왼쪽과 오른쪽 음영이 면적이 각각 0.05 인 위치 x1 과 x2 를 구하시오 
x1 = st.chi2.ppf(0.05,k)
print(x1)

x2 = st.chi2.ppf(0.95,k)
print(x2)

#왼쪽 음영과 오른쪽 음영의 면적이 각각 0.025 인 위치 x1과 x2 를 구하시오 
x1 = st.chi2.ppf(0.025,k)
x2 = st.chi2.ppf(0.975,k)
print(x1)
print(x2)
```

###6.스튜던트 t

```python
#스튜턴트 확률분포 97.5% 위치를 구하고 표준 정규 분포의 97.5% 분위수와 비교 하시오.

k = [5,20,100]
mu = 0
sigma = 1

for k in k:
  print(st.t.ppf(0.975,k))
print(st.norm.ppf(0.975,mu,sigma))
```
