```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/인공지능사관학교/빅데이터/05_통계와 파이썬_업데이트/data/data_iris.csv',header = 'infer',encoding = 'latin1')
```

```python
x = df['Petal.Length']
y = df['Sepal.Length']
```

```python
plt.scatter(x,y,color = 'green',alpha = 0.5,marker = 'o',s = 30)
plt.xlabel('Petal Length')
plt.ylabel('Sepal Length')
plt.title('IRIS FLOWER SCATTER PLOT')
plt.show()
```

#Pearson 상관계수

```python
np.round(st.pearsonr(x,y),3)
'''
return 값 
[피어슨 상관계수 값, p-value]
'''
```

```python
x.corr(y)
```

```python
np.round(df.corr(),3)
```

Spearman 상관계수

```python
np.round(st.spearmanr(x,y),3)
```
