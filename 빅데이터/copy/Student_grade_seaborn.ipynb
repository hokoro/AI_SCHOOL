## 학생 성적 데이터를 가지고 시각화 분석 



##columns
- gender : 성별	(남/여)
- race/ethnicity:그룹(명목형)
- parental level of education	:최종 학위(명목형)
- lunch	:점심
- test preparation course	시험 준비 과정
- math score	:수리 점수(수치)
- reading score	 : 읽기 점수 (수치)
- writing score : 쓰기 점수(수치)

```python
import seaborn as sns 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/StudentsPerformance.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

##histplot

```python
#연속형 변수의 기본 히스토 그램 
# 수학점수
sns.histplot(data = df ,kde = True , x = 'math score',bins = 10,color = 'purple')
plt.show()

print(df['math score'].mean()) ## 평균 66.089
```

```python
#reading score 
sns.histplot(data = df,x = 'reading score',bins = 10,color = 'green')
plt.show()
print(df['reading score'].mean())##69.169
```

```python
#writing score 
sns.histplot(data = df,x = 'writing score',bins = 10 ,color = 'red')
plt.show()
print(df['writing score'].mean()) ##68.054
```

#displot

```python
sns.displot(data = df,x = 'reading score',kde = True , rug = True ,bins = 10,color = 'blue',kind = 'hist')
```

```python
#kde - reading score
sns.displot(data = df,x = 'reading score',hue = 'race/ethnicity',kind = 'kde',aspect = 1.5,height = 3)
#group C 반 이 Density 가 가장 높다
```

```python
#kde - mathscore
sns.displot(data = df,x = 'math score',hue = 'race/ethnicity',kind = 'kde',aspect = 1.5,height = 3)
#group C 반 역시 density 가 가장 높음 
```

```python
#kde -writingscore
sns.displot(data = df,x = 'writing score',hue = 'race/ethnicity',kind = 'kde',aspect = 1.5,height = 3)
#group C 반 역시 density 가 가장 높음 
```

##산점도  + 회귀선 

```python
#jointplot - scatter
sns.jointplot(data = df,x = 'math score',y = 'reading score',color = 'red',kind = 'scatter')
```

```python
#jointplot - reg(회귀선 그리기)
sns.jointplot(data = df , x = 'writing',y = 'writing score',color = 'orange',kind = 'reg')
```

```python
fig  = sns.jointplot(data = df , x = 'reading score',y = 'writing score',color = 'green',kind = 'hex')
'''
hex 육각형 모델 로 표시 
확실히 60점 구간대가 제일 높은 빈도를 나타내고 있다. 
'''
```
