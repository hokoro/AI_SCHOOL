## 3 marker 찍기

- marker 생성
  
marker =   folium.Marker(location = [위도,경도] , popup = 마우스로 클릭했을때 (html 태그) , tooltip = 마우스가 올려질때 보이는 글 ) 

- marker 를 추가하는 방법 

marker.add_to(map) = marker 를 map 에 추가하는 방식 
map.add_child(marker) =  map 에 marker 를 자식으로 만드는 방식 

- Circle 원 추가하는 방식 

folium.Circle([위도 , 경도], popup = 마우스로 클릭했을때 (html 태그), radius = 원의 반지름, color='원의 색깔', fill_color = '원 안에 색깔', fill = 색깔의 채울것인지).add_to(my_map)


- CircleMarker 추가하는 방식 
folium.CircleMarker([위도 , 경도], popup = 마우스로 클릭했을때 (html 태그), radius = 반지름, color='원의 색깔', fill_color = '원 안에 색깔', fill = 색깔의 채울것인지).add_to(my_map)

- ClickForMarker 클릭 하면 지도 에 marker 가 표시되는 방식

my_map.add_child(folium.ClickForMarker(popup= html 태그)) 

- folium.LatLngPopup().add_to(my_map)  
클릭 하면 위도 경도를 표시 


```python
my_map = folium.Map(location = [35.229326,126.85968912],zoom_start=10,tiles = 'OpenStreetMap')
#my_map
my_marker = folium.Marker(location = [35.229326,126.85968912],popup = '<i>광주 AI</i>',tooltip = 'we are here!')
my_marker.add_to(my_map) #marker 를  map 에 추가하는 방식 
my_marker2 = folium.Marker(location = [35.229326,127] , popup = '<i>광주</i>',tooltip='Hello' )
my_map.add_child(my_marker2) #map 에서 child 로 marker 로 넣는 방식
my_map
```

```python
my_map = folium.Map(location=[35.229326,126.85968912], zoom_start=15, tiles='OpenStreetMap')
folium.Marker(location=[35.229326,126.85968912], popup='<b>Hello!</b>', tooltip='A', icon=folium.Icon(color='green',icon='cloud')).add_to(my_map)
folium.Marker(location=[35.229326,126.85968912+0.002], popup='<b>Gwangju</b>', tooltip='B', icon=folium.Icon(color='blue', icon='flag')).add_to(my_map)
folium.Marker(location=[35.229326,126.85968912+0.004], popup='<b>AI</b>', tooltip='C', icon=folium.Icon(color='red', icon = 'info-sign')).add_to(my_map)
folium.Marker(location=[35.229326,126.85968912+0.006], popup='<b>School</b>', tooltip='D', icon=folium.Icon(color='orange', icon = 'bookmark')).add_to(my_map)
my_map
```

```python
# 서클 세 개.
my_map = folium.Map(location=[35.229326,126.85968912], zoom_start=15, tiles='OpenStreetMap')
folium.Circle([35.229326,126.85968912], popup='<b>Hello!</b>', radius = 30, color='#FF0000', fill=False).add_to(my_map)
folium.Circle([35.229326,126.85968912+0.002], popup='<b>Gwangju</b>', radius = 60, color='#00FF00', fill_color = '#FF00FF', fill = True).add_to(my_map)
folium.Circle([35.229326,126.85968912+0.004], popup='<b>AI</b>', radius = 90, color = '#0000FF',fill=False).add_to(my_map)
my_map
```

```python
# 클릭해서 새로운 마커 추가.
# 더블 클릭해서 마커 삭제.
my_map = folium.Map(location=[35.229326,126.85968912], zoom_start=15, tiles='OpenStreetMap')
# folium.ClickForMarker(popup='<strong>Hello!</strong>').add_to(my_map)    # 방법 1
my_map.add_child(folium.ClickForMarker(popup='<strong>Hello!</strong>'))   # 방법 2.
my_map
```

```python
# 데이터 가져오기.
df = pd.read_csv('/content/drive/MyDrive/인공지능사관학교/빅데이터/04_시각화와 분석_업데이트 2/data/data_PatientRoute.csv')
```

```python
# MarkerCluster 붙이기.

#위도 경도 데이터 가져오기 
my_data = df[ ['latitude', 'longitude'] ]                    # DataFrame 또는 Numpy 배열.
my_popups = list(df['patient_id'])    #환자 동선에 보여질 환자 번호                        # List.
my_map = folium.Map([36.5, 128], zoom_start=7) #지도 만들고
plugins.MarkerCluster(locations=my_data, popups=my_popups).add_to(my_map)
#location = 위치 가져오기 / popups  = 환자 아이디 가져오기 
my_map
```

```python
# HeatMap 작성.
my_map = folium.Map(location=[36, 127], tiles='CartoDB Positron', zoom_start=7)
plugins.HeatMap(data=my_data, min_opacity = 0.3, radius=15).add_to(my_map)  # radius 에 민감하게 반응!
my_map
```
