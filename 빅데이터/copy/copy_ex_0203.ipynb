###확률 실험

```python
#확률 실험 
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np
my_list =[]


for _ in range(10000):
  z = np.random.randn()
  my_list.append(z)

```

```python
plt.hist(my_list,bins =100 , color = 'green',alpha = 0.3 , density = True)
plt.show()
```

```python
Z= np.array(my_list)
```

```python
np.round(Z.std(),2)
```

```python
#표준화 를 변환 X 에 대해서
mu = 10
sigma = 3
x = sigma * Z + mu

plt.hist(Z,bins =100 , color = 'green',alpha = 0.3 , density = True)
plt.hist(x,bins =100 , color = 'red',alpha = 0.3 , density = True)
plt.show()
```

```python
#카이제곱 
my_list = []

for _ in range(10000):
  Z1 = np.random.randn()
  Z2 = np.random.randn()
  Z3 = np.random.randn()
  Z4 = np.random.randn()
  Q = Z1 ** 2 + Z2 ** 2 + Z3 ** 2 + Z4 ** 2
  my_list.append(Q)

plt.hist(my_list,bins =100 , color = 'red',alpha = 0.3 , density = True)
plt.show()
```
