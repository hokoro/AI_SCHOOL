```python
from konlpy.tag import Okt   
okt=Okt()   
print(okt.morphs("멋쟁이사자처럼과 함께 코딩을 배워 보아요", stem=True))   # 어간추출 적용한 토큰화.
```

```python
print(okt.nouns("멋쟁이사자처럼과 함께 코딩을 배워 보아요"))  
```

```python
print(okt.pos("멋쟁이사자처럼과 함께 코딩을 배워 보아요"))  
```

```python
from konlpy.tag import Kkma   
kkma=Kkma()   
print(kkma.morphs("멋쟁이사자처럼과 함께 코딩을 배워 보아요")) 
```

```python
print(kkma.nouns("멋쟁이사자처럼과 함께 코딩을 배워 보아요"))   
```

```python
print(kkma.pos("멋쟁이사자처럼과 함께 코딩을 배워 보아요"))  
```

```python
from hanspell import spell_checker
```

```python
my_sentence = "철자법틀리면않돼나? 지금문장이상하다."
my_sentence_corrected = spell_checker.check(my_sentence).checked #모듈에서 틀린 맞춤법을 바로 잡아준다 
print(my_sentence_corrected)      
```

```python
from ckonlpy.tag import Twitter
```

```python
twi = Twitter()
```

```python
twi.morphs("멋쟁이사자처럼과 함께 코딩을 배워보아요")
```

```python
twi.pos("멋쟁이사자처럼과 함께 코딩을 배워보아요") #품사 태깅
```

```python
twi.add_dictionary('멋쟁이사자처럼','Noun') #본인이 넣고 싶은 단어를 품사 와 함께저장
```

```python
twi.morphs("멋쟁이사자처럼과 함께 코딩을 배워보아요") #분절
```

```python
twi.nouns("멋쟁이사자처럼과 함께 코딩을 배워보아요") #명사 추출
```
