## 분위수 를 계산 해보자 


```python
import scipy.stats as st
import numpy as np
```

##표준 정규 분포 

```python
#표준 정규 분포를 가자고 분위 수를 계산 해보자 
mu = 0 #평균
sigma = 1 #표준편차
q = 0.95  #확률 
a = st.norm.ppf(q,loc = mu , scale = sigma) # 그 지점의 분위수
 
st.norm.cdf(x = a ,loc = mu,scale = sigma ) # 그 지점의 퍼센테이지
```

```python
q = 0.975
b= st.norm.ppf(q,loc = mu,scale = sigma)

st.norm.cdf(x = b, loc = mu , scale = sigma )
```

##Student T 분포

```python
q = 0.95 
for df in [10,50,100,2000,10000]: # 자유도 의 배열 
  val = st.t.ppf(q,df)
  print('df = {} , 분위수  = {:.3f}'.format(df,val))

  #자유도가 올라가면 1.64 로 수렴 하는것을 확인 할수 있다.
```

```python
q = 0.975
for df in [10,50,100,2000,10000]:
  val = st.t.ppf(q,df) #q 랑 df 만 집어 넣도 값이 나오는것을 확인 평균과 표준편차가 정규 분포 default 를 따른다. 
  print('df = {} , 분위수  = {:.3f}'.format(df,val)) 
  #자유도 가 올라가면 1.96 으로 수렴 하는것을 확인 
```
