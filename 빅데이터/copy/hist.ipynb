```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np 
```

```python
df =  pd.read_csv('/content/drive/MyDrive/인공지능사관학교/빅데이터/data/data_studentlist_en.csv',header = 'infer',encoding = 'latin1')
```

## 분할표

```python
pd.crosstab(bloodType,gender)
```

```python
pd.crosstab(bloodType,gender,margins = True)
```

```python
pd.crosstab(bloodType,gender).plot.bar()
```
