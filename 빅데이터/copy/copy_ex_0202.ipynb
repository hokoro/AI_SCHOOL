```python
#연속 확률 분포 계산
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np
```

```python
#연속 균등 분포 함수
a = 2000
b = 5000
x = 3000
base = b-a  #가로변 
st.uniform.pdf(x,a,base) # 1 / 3000 의 값이랑 똑같음 
1 / (b-a)
```

```python
#분위수
st.uniform.ppf(0,a,base) #50% 확률이 있는 위치 0 %  = 2000 100% = 5000
```

```python
#누적확률 계산 
st.uniform.cdf(5000,a,base) #확률의 넓이로 계산 위치를 주면 확률을 반환 
```

```python
#구간 확률 계산 
st.uniform.cdf(3000,a,base) - st.uniform.cdf(2500,a,base)
```

```python
#확률 밀도 시각화 
x = np.linspace(1000,6000,1000) #x좌표 설정 
plt.plot(x,st.uniform.pdf(x,a,base),color = 'red')
plt.show()
```

###정규 분포 함수

```python
#확률 밀도 
mu = 5
sigma = 2
x = 0
st.norm.pdf(x,mu,sigma) #정규분포 확률 밀도 계산 
```

```python
#분위수 
st.norm.ppf(0.5,mu,sigma) #가운데 위치는 평균이 나와야 한다.
st.norm.ppf(-1,mu,sigma) # - inf
st.norm.ppf(1,mu,sigma) # inf
st.norm.ppf(0.7,mu,sigma)
```

```python
#누적확률 
st.norm.cdf(5,mu,sigma) #mu = 5 기 때문에 확률로는 절반이 나온다.
st.norm.cdf(6.048801025416081,mu,sigma)
```

```python
#구간의 확률 
st.norm.cdf(7,mu,sigma) - st.norm.cdf(3,mu,sigma)
```

```python
#시각화 
x = np.linspace(-10,10,1000)
mu = [-1,0,1]
color = ['red','blue','green']
sigma = 1
for m,c in zip(mu,color):
  plt.plot(x,st.norm.pdf(x,m,sigma),color = c)
plt.show()
```

```python
x = np.linspace(-10,10,1000)
mu = 0
color = ['red','blue','green']
sigma = [1,2,3]
for s,c in zip(sigma,color):
  plt.plot(x,st.norm.pdf(x,mu,s),color = c)
plt.show()
```

###카이 제곱 분포

```python
#확률 밀도 
df = 5      #degree of freedom(자유도)
x = 2

st.chi2.pdf(x,df)
```

```python
#분위수 
st.chi2.ppf(0.5,df) #50% 중위수
```

```python
#누적확률
st.chi2.cdf(4.351460191095526,df = 5)
```

```python
#시각화 
x = np.linspace(0,15,2000)
df = [1,5,30]
color = ['red','blue','green']
for d,c in zip(df,color):
  plt.plot(x,st.chi2.pdf(x,df = d),color = c)
plt.show()
```
