```python
import numpy as np 
import matplotlib.pyplot as plt 
import scipy.stats as st
```

```python
#표준 편차 가 2 이고 표본 평균이 3 인것 n = 10
np.random.seed(1234)
x = np.random.randn(10) *2 +3
print(len(x)) #표본의 크기 
```

```python
mu_0 = 0
st.ttest_1samp(x,mu_0)
#유의 수준 5% 라고 할때 pvalue 가 작기때문에 대립가설이 선택 되는 상황 (양측검정)
```

```python
#검정 통계량
mu_0 = 0
n = len(x)
x_mean = x.mean()
SEM = x.std(ddof=1)/np.sqrt(n)
(x_mean - mu_0)/SEM
```

```python
#pvalue 구하기 
mu_0 = 0
n = len(x)
x_mean = x.mean()
SEM = x.std(ddof = 1) / np.sqrt(n)
(1-st.t.cdf(np.abs((x_mean-mu_0)/SEM),df=n-1))*2
```

```python
mu_0 = 1
n = len(x)
x_mean = x.mean()
SEM = x.std(ddof =1) /np.sqrt(n)
(1-st.t.cdf(np.abs((x_mean-mu_0)/SEM),df=n-1))*2
#귀무가설 기각 0.037 < 0.05
```

##모집단이 2개 이상 

```python
x1 = np.random.randn(10)*2+3
x2 = np.random.randn(20)*2+3

n1 = len(x1)
n2 = len(x2)

x1_mean = x1.mean()
x2_mean = x2.mean()
s1=  x1.std(ddof = 1)
s2 = x2.std(ddof = 1)
```

```python
#표본을 직접 넣어 가진 독립표본 T 검정 
st.ttest_ind(x1,x2,equal_var=True)
```

```python
'''
기술 통계에서 두 개의 독립 표본 평균에 대한 T-검정.

이것은 두 개의 독립 표본이 동일한 평균(예상) 값을 갖는다는 귀무 가설에 대한 양측 검정입니다.
'''
st.ttest_ind_from_stats(x1_mean,s1,n1,x2_mean,s2,n2,equal_var = True)
```

```python
'''
등분산 검정을 실시한다. 
'''
st.bartlett(x1,x2)
```

```python
x1 = np.random.randn(10) * 5 + 3
x2 = x1 + np.random.randn(10) * 1 + 3
#대응표본 검정 함수 (양측검정만 됨)
st.ttest_rel(x1,x2) #표본이 10개 대응 pvalue 가 너무 작아서 귀무가설 기각 대립가설 채택
```
