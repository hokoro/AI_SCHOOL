##확률분포 

```python
import scipy.stats as st  #scipy 에 stats(통계) 
import matplotlib.pyplot as plt 
import numpy as np 
%matplotlib inline 


```

####이항 확률 분포 

```python
n = 10 #시행횟수 
p = 0.5 #확률 
st.binom.pmf(5,n,p)
```

```python
st.binom.ppf(0.3,n,p) #이항 분포 함수의 데이터가 어느 위치에 있는지 
```

```python
#누적 확률 계산 
#P(0<=x<=5)

st.binom.cdf(5,n,p) # cdf 는 누적확률 계산 함수 

```

```python
#시작지점이 0 이 아닐떄 구간 확률 

st.binom.cdf(7,n,p) - st.binom.cdf(2,n,p)
'''
0~ 7 인 누적확률에서 
0~ 2 인 누적확률을 뺴면 
3~ 7 사이의 구간의 누적확률을 계산 할수 있다.
'''
```

```python
x = np.arange(0,11)
plt.scatter(x/10,st.binom.ppf(x/10.0,n,p),color = 'red')
plt.show()
```
